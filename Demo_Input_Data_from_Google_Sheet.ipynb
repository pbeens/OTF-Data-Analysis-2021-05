This set of data is fairly simple to process and doesn't require any repair. 

The data is coming from a Google Sheet that has had its permission set to anyone with the link being able to view.

Our goal is to read the data into a dataframe and plot it as a line graph.

```python
# Read in pandas library

import pandas as pd
```

```python
# Make sure the Google Sheet is shared with anyone with the link.
# Note the last part of the URL. It must be changed to /export?format=csv
# The original URL is https://docs.google.com/spreadsheets/d/1UmY4bodvLONJLTPrWH4gyKPMhFMuuwSnr4rJ3VSaK7k/edit#gid=0

url = 'https://docs.google.com/spreadsheets/d/1UmY4bodvLONJLTPrWH4gyKPMhFMuuwSnr4rJ3VSaK7k/export?format=csv'
```

```python
# Read in the data from the Google Sheet and put it in a dataframe

df = pd.read_csv(url)
```

```python
# Let's look at the first few rows

df.head()
```

```python
# What are the column names?

df.columns
```

```python
# Let's plot the data using pandas
# Some options to play with here: grid, title

df.plot(kind='line', 
        x='X', 
        y='Y',
        ylabel='$X^2$', # the $$ enables TeX markup (^, _, frac)
        legend=False)
```

Here are some other examples of using matplotlib instead of pandas to plot the data.

```python
import matplotlib.pyplot as plt
```

```python
# x-y plot

# map the columns to x, y for simplicity
x = df.X
y = df.Y

# x-y plot
plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('$X^2$')
plt.show()
```

```python
# scatter plot

plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('$X^2$')
plt.show()
```
